### **Multiplicación**

$$\alpha = pendiente$$

$$\beta = interseccion$$

$$n = observaciones$$

$$y ═ precio$$

$$x ═ fecha$$


La solución para la pendiente es:

$$\alpha = \frac{n\sum xy - \sum x \sum y}{n \sum x^2 - (\sum x)^2}$$


La solución para la intersección es:

$$\beta = \frac{\sum y - \alpha \sum x}{n}$$


La solución para la ecuación de la recta es:

$$y = \alpha x + \beta$$



```python
n = np.size(moscu)
y = moscu
x = fechas
```

$$\alpha = \frac{n\sum xy - \sum x \sum y}{n \sum x^2 - (\sum x)^2}$$

```python
a = ((n*np.sum(x*y))-(np.sum(x)*np.sum(y)))/(n*np.sum(x**2)-(np.sum(x)**2))
```

$$\beta = \frac{\sum y - \alpha \sum x}{n}$$

```python
b = np.mean(y) - a * np.mean(x) # mean para promedios
```

```python
print(f'La pendiente es: {a}\nLa intersección es: {b}')
```

$$y = \alpha x + \beta$$

```python
y = (a*x)+b
```

### Tercer parte

Ahora debes calcular el coeficiente angular y lineal para la recta de las naranjas y para la recta de las toronjas. Utiliza la fórmula de mínimos cuadrados para encontrar cada uno.

```python
n_y = pesos_naranjas
n_x = diametros_naranjas
n_n = np.size(n_x)
```

$$\alpha = \frac{n\sum xy - \sum x \sum y}{n \sum x^2 - (\sum x)^2}$$

```python
n_a = ((n_n*np.sum(n_x*n_y))-(np.sum(n_x)*np.sum(n_y)))/((n_n*np.sum(n_x**2))-(np.sum(n_x))**2)
```

$$\beta = \frac{\sum y - \alpha \sum x}{n}$$

```python
n_b = np.mean(n_y) - n_a * np.mean(n_x)
```

```python
n_y = (n_a*n_x)+n_b
```

```python
n_v = np.linalg.norm(pesos_naranjas - n_y)
```

```python
print(f'La pendiente es: {n_a}\nLa intersección es: {n_b}\nLa norma del vector es: {n_v}')
```

```python
# Grafico
plt.title('Peso en funcion del diametro', fontsize=16, weight='bold')
plt.xlabel('Diametro', fontsize=12)
plt.ylabel('Peso', fontsize=12)

plt.plot(diametros_naranjas,pesos_naranjas, color='orange', label = "Naranjas")
plt.plot(n_x,n_y, color='skyblue', label = "Regresion lineal")
plt.legend()
plt.show()
```
